# Creating SDP of sin waves


```python
import numpy as np
import matplotlib.pyplot as plt
time = np.arange(0,10,1/2400)
amp = np.sin(2*np.pi*time*50)
fig = plt.figure()
fig.set_size_inches(12,6)
plt.subplots_adjust(hspace=1)

plt.subplot(311)
# plt.plot(x, sampled_f1)
plt.plot(time,amp)
plt.xlim(0, 1)
```

```python
# SDP OF SIN WAVE
for i in range(1,7):
    rs = (amp - np.min(amp))/(np.max(amp)- np.min(amp))
    th = (60*i + ((amp - np.min(amp))/(np.max(amp)- np.min(amp))*35))
    ph = (60*i - ((amp - np.min(amp))/(np.max(amp)- np.min(amp))*35))
    th = th[2:]
    ph = ph[2:]
    rs = rs[:len(rs)-2]
    plt.polar(th*(np.pi/180),rs,'.', c='blue')
    plt.polar(ph*(np.pi/180),rs,'.', c='red')
```
